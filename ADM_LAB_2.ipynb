```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Titanic.csv')
data.head()
```

```python
data.describe()
```

```python
data.describe(include=['object'])
```

```python
import matplotlib.pyplot as plt

plt.hist(data['age'].dropna(), bins=20, edgecolor='black', color='red')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['pclass'].value_counts().plot(kind='bar', color='blue')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['age'].dropna(), patch_artist=True)
boxprops=dict(facecolor='blue')
plt.title('Box plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns

sns.pairplot(data[['age', 'fare', 'sibsp', 'parch']].dropna())
plt.show()
```

```python
correlation_matrix = data[['age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=data, palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.show()
```
