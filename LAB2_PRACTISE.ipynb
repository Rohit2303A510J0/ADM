```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.array([0, 6]), np.array([0, 250]), 3)
plt.show()
```

```python
plt.plot(np.array([3, 8, 1, 10, 5, 7]), mec='royalblue', mfc='red')
plt.show()
```

```python
plt.plot(np.array([3, 8, 1, 10, 5, 7]), 'o:r')
plt.show()
```

```python
plt.plot(np.array([3, 8, 1, 10, 5, 7]), 'd:b')
plt.show()
```

```python
plt.plot(np.array([3, 8, 1, 10, 5, 7]), ls='solid', linewidth = '5')
plt.show()
```

```python
plt.plot(np.array([0, 1, 2, 3]), np.array([3, 8, 1, 10]), np.array([0, 1, 2, 3]), np.array([6, 2, 7, 11]))
plt.show()
```

```python
plt.bar(np.array(['A', 'B', 'C', 'D']), np.array([3, 8, 1, 10]))
plt.title('Bar Chart Example')
plt.xlabel('Bar Number')
plt.ylabel('Bar Height')
plt.show()
```

```python
plt.scatter(np.array(['A', 'B', 'C', 'D']), np.array([3, 8, 1, 10]))
plt.title('Scatter Example')
plt.show()
```

```python
plt.hist(np.random.normal(170, 10, 250))
plt.show()
```

```python
plt.pie(np.array([35, 25, 25, 15]))
plt.show(0)
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([2, 3, 4, 6])

plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.show()

x2 = np.array([3, 4, 5])
y2 = np.array([5, 6, 7])
plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([2, 3, 4, 6])
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.show()
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0.1, 0, 0.1]
plt.pie(y, labels = mylabels, explode=myexplode)
plt.show()
```
